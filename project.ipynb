```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv('old_car_data.csv')
```

```python
dataset.head(4)
```

```python
car_ID_le=LabelEncoder()
dataset['car_ID']=car_ID_le.fit_transform(dataset["car_ID"])
```

```python
car_Name_le=LabelEncoder()
dataset['CarName']=car_Name_le.fit_transform(dataset["CarName"])
```

```python
fuel_type_le=LabelEncoder()
dataset['fueltype']=fuel_type_le.fit_transform(dataset["fueltype"])
```

```python
input_data=dataset.iloc[:,:-1]
output_data=dataset["price"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data),columns = input_data.columns)
```

```python
from sklearn.model_selection import train_test_split
dataset.head(5)
```

```python
x_train,x_test,y_train,y_test =train_test_split(input_data,output_data,train_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators = 100)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)*100, rf.score(x_test,y_test)*100
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
mean_squared_error(y_test,rf.predict(x_test)),mean_absolute_error(y_test,rf.predict(x_test))
```
